```python
import pandas as pd

data = pd.read_csv("iris.data",names=['a','b','c','d','label'])
data
```

```python
data = []

for line in open('iris.data','r'):
  raw = line.split(",")
  x = [float(r) for r in raw[:4]]
  y = raw[4].strip()
  data.append((x,y))
data
```

```python
import matplotlib.pyplot as plt
X = []
Y = []

for v,l in data:
  X.append(v[0])
  Y.append(v[1])
classes = set(l for v,l in data)
print(classes)

plt.scatter(X,Y,alpha = 0.5)
plt.show()
```

```python
classes = {l:[] for v,l in data}
for v,l in data:
  classes[l].append(v)

print(classes)
for c, c_data in classes.items():
  X=[]
  Y=[]
  for v in c_data:
    X.append(v[2])
    Y.append(v[3])
  plt.scatter(X,Y,alpha = 0.5)
plt.show()
```

```python
def distance(a,b):
  s = 0
  for x,y in zip(a,b):
    s += (x-y)**2
  
  return s**0.5
```

```python
from collections import Counter

def knn_classify(k,query,data):

  while True:
    topk = sorted((distance(query,v),l)for v,l in data)[:k]
    cnts = Counter(l for d,l in topk)
    mc = cnts.most_common(2)
    if len(cnts) == 1 or mc[0][1] != mc[1][1]:
      break
    else:
      k -= 1
  
  return cnts.most_common(1)[0][0]

k = 3
res = knn_classify(k,[6.4,2.8,5.6,2.2],data)
print(res)
```

```python
from matplotlib import pyplot as plt

x = []
y = []
for k in range(1,50):
  n_correct = 0
  n_incorrect = 0
  for i ,(v,l) in enumerate(data):
    new_data = []
    for j,d in enumerate(data):
      if i == j: continue
      new_data.append(d)
    ans = knn_classify(k,v,new_data)
    if l == ans:
      n_correct += 1
    else:
      n_incorrect += 1

  accuracy = n_correct/(n_correct+n_incorrect)
  x.append(k)
  y.append(accuracy)
  print(accuracy)

plt.plot(x,y)
plt.show()

```
